```python
import pandas as pd
```

```python
df = pd.read_csv("./datasets/insurance-marketing-dataset.csv", delimiter=";")
df
```

```python
all_columns = list(df.columns)
print(all_columns)
```

- Active holding of coverage for nine different categories of products:     
  (first year / time zero) `HAS_A_1`, `HAS_B_1`, `HAS_C_1`, ..., `HAS_I_1`     
  (second year / time one) `HAS_A_2`, `HAS_B_2`, `HAS_C_2`, ..., `HAS_I_2`
- Fifteen categorical and 14 numerical features representing the policyholder's    
  demographic profile measured in the first year only    
  `DEMO_CAT_01`, `DEMO_CAT_02`, ...., `DEMO_CAT_14`, `DEMO_CAT_15`    
  `DEMO_NUM_01`, `DEMO_NUM_02`, ...., `DEMO_NUM_14`   
- The policyholder unique ID    
  `UNIQUE_ID`
- The condition that is indicating whether the policyholder acquired the product   
  between the first year and second year. For example    
  `BUYED_E` = (`HAS_E_1 == 0` OR `HAS_E_2 == 1`)
